# SUMMERTIME
Este feature se consigue a través de tomar a todas las propiedades con pileta que se encuentren al sur del Trópico de Cáncer, ya que se encontrarían en la Zona intertropical, donde la variación de temperatura es muy baja, y hace calor todo el año. Entonces, todas las propiedades que se encuentren en este grupo, tendrán un valor más elevado.

```python
#tropico de cancer = a latitud  23,43
df = df.dropna(subset=['lat', 'piscina'])

df['calurosa'] = False
df.loc[((df['lat'] < 23.43)) & ((df['piscina'] > 0)) , 'calurosa'] = True

df.sort_values('piscina')
```

# Escuelas y cantidad de habitaciones
Si una casa tiene muchas habitaciones, es probable que sea para que haya chicos, por lo que tener escuelas cercanas va a ser de gran utilidad

```python
df = df.dropna(subset=['habitaciones', 'escuelascercanas'])

df['parachicos'] = False
df.loc[((df['habitaciones'] > 2)) & ((df['escuelascercanas'] > 0)) , 'parachicos'] = True

df.sort_values('habitaciones')
```
